# Defination 

A support vector machine (SVM) is a type of supervised learning algorithm used in machine learning to solve classification and regression tasks; SVMs are particularly good at solving binary classification problems, which require classifying the elements of a data set into two groups.

Idea behind SVM is create a separating hyperplane between the classes.

We could use the separator that maximizes the margins between the classes. This is called as the Maximal Margin Classifire.  

# Hyperplanes
The SVM algorithm helps to find the best line or decision boundary; this best boundary or region is called as a hyperplane.
# Margin
The distance between the vectors and the hyperplane is called as margin

# Kernels
Kernels allows us to move beyond a Support Vector machines.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mouse_viral_study.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='Med_1_mL', y = 'Med_2_mL', hue ='Virus Present',
               data =df)
# HYPERPLANE (2D 0line)

x= np.linspace(0,10,100)

m = -1
b=11
y = m*x+b
plt.plot(x,y,'black')
```

```python
from sklearn.svm import SVC
```

```python
# help(SVC)
```

```python
y= df['Virus Present']
```

```python
X = df.drop('Virus Present',axis = 1)
```

```python
model = SVC(kernel='linear', C = 1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='linear',C=0.005)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C = 1,gamma=0.9)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model =SVC(kernel='sigmoid')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model  = SVC(kernel='poly',C =1,degree=2)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm =SVC()
param_grid ={'C':[0.01,0.1,1],'kernel':['linear','rbf']}
```

```python
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```
